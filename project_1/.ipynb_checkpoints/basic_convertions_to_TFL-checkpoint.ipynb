```python
import tensorflow as tf
```

```python
x = [-1, 0, 1, 2, 3, 4]
y = [-3, -1, 1, 3, 5, 7]
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(x, y, epochs = 500)
```

```python
print(model.predict([10.0]))
```

```python
import pathlib
```

```python
export_dir = "./saved_model"
tf.saved_model.save(model, export_dir)
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(export_dir)
tflite_model = converter.convert()
```

```python
tflite_model_file = pathlib.Path('TFLiteModel/hello_world.tflite')
tflite_model_file.write_bytes(tflite_model)
```

```python
############ -- converting a existing pretrained mobilenet to tflite -- ###############
```

```python
model2 = tf.keras.applications.MobileNetV2(weights="imagenet", input_shape=(224, 224, 3))
model2.save('model2.h5')
```

```python
converter2 = tf.lite.TFLiteConverter.from_keras_model(model2)
tflite_model2 = converter2.convert()
```

```python
tflite_model_file2 = pathlib.Path('TFLiteModel/mobilenet.tflite')
tflite_model_file2.write_bytes(tflite_model2)
```
